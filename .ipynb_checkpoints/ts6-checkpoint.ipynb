# I. **Introducción**

## Función de Transferencia $ H(S) $
Un sistema lineal e invariante en el tiempo (*LTI*) se puede caracterizar por su **función de transferencia** $ H(s) $. La misma es una expresión matemática que caracteriza las relaciones de *“Entrada – Salida”* de los sistemas. Se define como la relación de la transformada de Laplace de la salida (*función respuesta*), a la transformada de Laplace de la entrada (*función excitadora*). Dentro de este dominio la variable es $ s $, pertenece a los complejos y se define como $s = \theta +j\omega$. Trabajar en este dominio permite un análisis transitorio del régimen permanente, es decir, el análisis temporal.

$$
H(s) = \frac {Y(s)}{X(s)}
$$

O lo que es análogo, $Y(s) = H(s)X(s)$

## Características de un sistema

1. Si se tiene un sistema de segundo orden, por ejemplo un circuito *RLC*, existe una característica de dichos sistemas que se denomina **frecuencia natural o de resonancia** y viene dada por: $ \omega_0 = \frac {1}{\sqrt{LC}} $. 

2. El **factor de calidad** $ Q $ de un filtro es una medida de la agudeza (selectividad) de la magnitud de la función de transferencia. Para un circuito serie RLC su forma es: 
$$
Q = \frac {1}{R} \sqrt {\frac {L}{C}}
$$

Un valor alto de $ Q $ implica una curva de magnitud más aguda, y por lo tanto, mayor selectividad.

## Polos y ceros

- Los **polos** de la función $ H(s) $ son las raíces del denominador y determinan la estabilidad y la forma de respuesta transitoria del sistema.
- Los **ceros** son las raíces del numerador y modifican la respuesta en frecuencia atenuando o anulando ciertos componentes espectrales.

Para un denominador genérico de segundo orden: 
$$ s^2 + \frac{\omega_{0z}}{Q_z} s + \omega_0^2 $$

los polos vienen dados por: 

$$
s_p = -\frac{\omega_{0p}}{2Q_p} \pm j\omega_{0p} \sqrt{1 - \frac{1}{4Q_p^2}}
$$

Según el valor de $Q_p$:
- Si $ Q_p > 1/2 $ los polos son complejos conjugados (resonancia con amortiguamiento).
- Si $ Q_p = 1/2 $ los polos son reales iguales (amortiguamiento crítico).
- Si $ Q_p < 1/2 $ los polos son reales distintos (sobreamortiguado).

Los ceros se calculan de forma análoga para el numerador. 

## Respuesta en frecuencia - Diagrama de fase y respuesta de módulo
- La **respuesta en magnitud $ |H(j\omega)| $** muestra cómo se atenuan o amplifican distintas frecuencias.
- La **respuesta de fase $ \angle H(j\omega) $** indica el desfase introducido entre la entrada y la salida.

En los diagramas de Bode:
- Un **polo simple** introduce una pendiente adicional de aproximadamente $ -20 dB/decada $.
- Un **cero simple** aporta $ +20 dB/decada $.

Las asíntotas permiten obtener una aproximación rápida, corregida luego en torno a las frecuencias de quiebre (desvíos de alrededor de $\pm 3 dB$ en $\omega_0$).

### Retardo de grupo
El **retardo de grupo** $ \tau_g(\omega) $ cuantifica la distorsión temporal de envolventes y se define como: 

$$
\tau_g(\omega) = -\frac{d}{d\omega} \angle H(j\omega)
$$

Un retardo de grupo constante implica menor distorsión para señales de banda angosta. 

## Sección bicuadrática

Una **sección bicuadrática** (*Second-Order Section, SOS*) es un bloque de un sistema cuya función de transferencia tiene la forma de un cociente entre dos polinomios cuadráticos en la variable *s*. Es decir, es una estructura que representa un sistema de segundo orden (o la combinación de dos sistemas de segundo orden en cascada). La fórmula matemática que lo describe es:

$$
T(s) = k \cdot \frac{s^2 + \dfrac{\omega_z}{Q_z}s + \omega_z^2}{s^2 + \dfrac{\omega_p}{Q_p}s + \omega_p^2}
$$

Donde $ k $ representa la ganancia del sistema, $\omega_z$ y $\omega_p$ las frecuencias naturales asociadas a los ceros y polos, $Q_z$ y $Q_p$ los factores de calidad. El nombre del bloque se debe a que el numerador y el denominador son polinomios de segundo grado.

### Demostración de que $T(s)$ es una sección bicuadrática (SOS)

A partir del circuito mostrado en la Figura

<p style="text-align: center;"><em>Figura I: Circuito analizado correspondiente a la sección bicuadrática (SOS).</em></p>

la función de transferencia del sistema se define como:

$$
\begin{equation}
    T(s) = \frac{V_2(s)}{V_1(s)}.
\end{equation}
\quad (1)
$$

Considerando el análisis del circuito en el dominio de Laplace y las impedancias complejas de los componentes ($Z_R = R$, $Z_L = sL$, $Z_C = \frac{1}{sC}$), se obtiene que la relación entre la tensión de salida y la de entrada puede escribirse en función de polinomios de segundo orden en la variable compleja $s$:

<p style="text-align: center;">
  \( T(s) = k \cdot \frac{s^2 + \frac{\omega_{0Z}}{Q_Z}s + \omega_{0Z}^2}{s^2 + \frac{\omega_{0P}}{Q_P}s + \omega_{0P}^2} \quad (2) \)
</p>

Esta ecuación corresponde a la forma canónica de una *sección bicuadrática* o *Second Order Section (SOS)*, que describe un sistema de segundo orden caracterizado por:

- Un par de **ceros complejos conjugados**, determinados por $\omega_{0z}$ y $Q_z$.
- Un par de **polos complejos conjugados**, definidos por $\omega_{0p}$ y $Q_p$.
- Una **ganancia** global $k$ que determina la amplitud relativa de la respuesta.

Por lo tanto, la función de transferencia $T(s)$ obtenida a partir del circuito puede expresarse en la forma de una *sección de segundo orden*, verificando que se trata efectivamente de una sección bicuadrática.


## *<u> Punto 2 </u>*

### Parametrización de los polinomios de segundo orden

Para el circuito analizado, los polinomios del numerador y del denominador de la ecuación *(2)* pueden expresarse en términos de los elementos pasivos del circuito y de los coeficientes de división $a$, $b$ y $d$, según la estructura de la red mostrada en la Figura *I*.

Luego de realizar el análisis de nodos y la reducción de impedancias, se obtiene la siguiente expresión general:

$$
    T(s) = k \cdot
    \frac{s^2 + \frac{R_Z}{L_Z}s + \frac{1}{L_Z C_Z}}
         {s^2 + \frac{R_P}{L_P}s + \frac{1}{L_P C_P}}.
         \quad (3)
$$

A partir de la ecuación *(3)*, se identifican los parámetros característicos de la sección bicuadrática como:

$$
\begin{align}
    \omega_{0Z} &= \frac{1}{\sqrt{L_Z C_Z}}, &
    Q_Z &= \frac{1}{R_Z} \sqrt{\frac{L_Z}{C_Z}}, \\[6pt]
    \omega_{0P} &= \frac{1}{\sqrt{L_P C_P}}, &
    Q_P &= \frac{1}{R_P} \sqrt{\frac{L_P}{C_P}}.
\end{align}
$$

La ganancia $k$ depende de la relación de división de tensiones del circuito resistivo–capacitivo y puede escribirse como:

$$
\begin{equation}
    k = \frac{N_0}{D_0}.
\end{equation}
$$

Finalmente, considerando los elementos definidos en la red:

$$
\begin{align}
    L_Z &= \frac{L}{b}, & C_Z &= C \cdot a, & R_Z &= \frac{R}{d}, \\[6pt]
    L_P &= \frac{L}{1 - b}, & C_P &= C (1 - a), & R_P &= \frac{R}{1 - d},
\end{align}
$$

los parámetros de la sección bicuadrática pueden escribirse explícitamente en función de los componentes físicos del circuito:

$$
\begin{align}
    \omega_{0Z} &= \sqrt{\frac{b}{L C a}}, &
    Q_Z &= \frac{d}{R} \sqrt{\frac{L b}{C a}}, \\[6pt]
    \omega_{0P} &= \sqrt{\frac{1 - b}{L C (1 - a)}}, &
    Q_P &= \frac{1 - d}{R} \sqrt{\frac{L (1 - b)}{C (1 - a)}}.
\end{align}
$$

De este modo, se logra parametrizar completamente la función de transferencia $T(s)$ en función de los componentes de la red, demostrando la correspondencia entre la topología del circuito y la forma canónica de una sección bicuadrática.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Hacemos un filtro pasa bajos

# Plantilla de diseño
wp = 1 # Frecuencia de corte/paso (rad/seg)
ws = 5 # Frecuencia de stop/detenida (rad/seg)

alpha_p = 1 # Atenuacion maxima a la wp, alfa_max, perdidas en banda de paso
alpha_s = 40 # Atenuacion minima a la ws, alfa_min, minima atenuacion requerida en banda de paso

# Aproximaciones de modulo
f_aprox_butter = 'butter'

# Diseño del filtro analogico
b, a = signal.iirdesign(wp = wp, ws = ws, gpass = alpha_p, gstop = alpha_s, analog = True, ftype = f_aprox_butter, output = 'ba')

# Matriz SOS
sos = signal.tf2sos(b, a, analog = True) # Ya me lo da en forma monica
# La raiz del ultimo valor del array me tiene que dar el radio 

print("\n\t\t\t\tMatriz SOS\n\n", sos); # Compruebo que es una seccion bicuadratica 
```

A partir de la matriz de *secciones de segundo orden* obtenida: 

$$
\text{SOS} =
\begin{bmatrix}
0 & 0 & 1.96522673 & 1 & 2.18775410 & 1.40186545\\[4pt]
0 & 0 & 1.00000000 & 1 & 0.90619742 & 1.40186545
\end{bmatrix}
$$

cada fila representa una sección bicuadrática de la forma general: 

$$
T(s) = \frac{b_0 + b_1 s + b_2 s^2}{a_0 + a_1 s + a_2 s^2}
$$

Normalizando el denominador para que el coeficiente de $s^2$ en el denominador sea 1 (dividiendo numerador y denominador por $a_2$), se obtiene la forma canónica:

$$
T(s) = k \cdot \frac{s^2 + \frac{\omega_{0Z}}{Q_Z}s + \omega_{0Z}^2}{s^2 + \frac{\omega_{0P}}{Q_P}s + \omega_{0P}^2}
$$

donde:

$$
k = \frac{b_2}{a_2}
$$

$$
\frac{\omega_{0P}}{Q_P} = \frac{a_1}{a_2}
$$

$$
\omega_{0P}^2 = \frac{a_0}{a_2}
$$

De estas igualdades se siguen las expresiones útiles

$$
\begin{align}
\omega_{0P} &= \sqrt{\frac{a_0}{a_2}} \\
Q_P &= \frac{\omega_{0P}}{a_1 / a_2} = \frac{\sqrt{a_0 a_2}}{a_1}
\end{align}
$$

- Polos: a partir del polinomio sin normalizar:

$$
a_2 s^2 + a_1 s + a_0 = 0
$$

La solución general es:

$$
s_p = \frac{-a_1 \pm \sqrt{a_1^2 - 4 a_2 a_0}}{2 a_2}
$$

- Ceros: si el numerador es $b_2 s^2$ (es decir, $b_0 = b_1 = 0$), entonces hay ceros dobles en: $s = 0$ (dos ceros en el origen).

Esto explica el comportamiento pasa-altos (al menos del tipo que cancela DC).

### **Primera sección** 

$$
[b_0,\, b_1,\, b_2,\, a_0,\, a_1,\, a_2] =
[0,\; 0,\; 1.96522673,\; 1,\; 2.18775410,\; 1.40186545]
$$

Al normalizar por $ a_2 = 1.40186545 $: 

$$
T_1(s) = k_1 \cdot
\frac{s^2}{s^2 + 1.5606\,s + 0.7133}
$$

dónde 

$$
k_1 = \frac{b_2}{a_2} = \frac{1.96522673}{1.40186545} = 1.4019.
$$

A partir de los coeficientes del denominador se obtiene: 

$$
\omega_{0P1} = \sqrt{0.7133} = 0.8446\ \text{rad/s}, \qquad
\frac{\omega_{0P1}}{Q_{P1}} = 1.5606
\Rightarrow
Q_{P1} = \frac{0.8446}{1.5606} = 0.5412.
$$

Los polos complejos conjugados correspondientes son: 

$$
s_{p1,2} = -0.7803 \pm j\,0.3232
$$

mientras que el numerador presenta un término proporcional a $s^2$, lo que implica la existencia de **dos ceros en el origen** $s = 0$. 

### **Segunda sección** 

$$
[b_0,\, b_1,\, b_2,\, a_0,\, a_1,\, a_2] =
[0,\; 0,\; 1.0000,\; 1,\; 0.90619742,\; 1.40186545]
$$

Normalizando nuevamente por $a_2 = 1.40186545 $: 

$$
T_2(s) = k_2 \cdot
\frac{s^2}{s^2 + 0.6464\,s + 0.7133}
$$

con

$$
k_2 = \frac{b_2}{a_2} = \frac{1.0000}{1.40186545} = 0.7133
$$

Para esta sección: 

$$
\omega_{0P2} = \sqrt{0.7133} = 0.8446\ \text{rad/s}, \qquad
\frac{\omega_{0P2}}{Q_{P2}} = 0.6464
\Rightarrow
Q_{P2} = \frac{0.8446}{0.6464} = 1.3066
$$

Los polos se ubican en: 

$$
s_{p1,2} = -0.3232 \pm j\,0.7803
$$

y al igual que en la sección anterior, el numerador posee **ceros dobles en el orien**. 

### Resumen de los parámetros hallados

| Sección | $k$ | $\omega_{0p}$ [rad/s] | $Q_p$ | Ceros | Polos $(s = \sigma \pm j\omega_0)$ |
|:-------:|:-----:|:-----------------------:|:-------:|:------:|:---------------------------------:|
| 1 | 1.4019 | 0.8446 | 0.5412 | $s = 0$ (doble) | $-0.7803 \pm j 0.3232$ |
| 2 | 0.7133 | 0.8446 | 1.3066 | $s = 0$ (doble) | $-0.3232 \pm j 0.7803$ |

De este modo, se obtienen numéricamente los parámetros que confirman la correspondencia entre la matriz *SOS* y la estructura teórica de una sección bicuadrática. 

```python
# %% PUNTO 3, 4 y 5
# Defino las tres funciones de trasferencia que me da la consigna
# T1(s) = (s^2 + 9) / (s^2 + sqrt(2)s + 1)
num1 = [1, 0, 9]
den1 = [1, np.sqrt(2), 1]
T1 = signal.TransferFunction(num1, den1)

# T2(s) = (s^2 + 1/9) / (s^2 + (1/5)s + 1)
num2 = [1, 0, 1/9]
den2 = [1, 1/5, 1]
T2 = signal.TransferFunction(num2, den2)

# T3(s) = (s^2 + 1/5 s + 1) / (s^2 + sqrt(2)s + 1)
num3 = [1, 1/5, 1]
den3 = [1, np.sqrt(2), 1]
T3 = signal.TransferFunction(num3, den3)

# Respuesta en frecuencia 
w = np.logspace(-2, 2, 2000) # Escala logaritmica 
w1, h1 = signal.freqs(num1, den1, w) # Calcula la respuesta en frecuencia del filtro
w2, h2 = signal.freqs(num2, den2, w) # Calcula la respuesta en frecuencia del filtro
w3, h3 = signal.freqs(num3, den3, w) # Calcula la respuesta en frecuencia del filtro

# Cálculo de fase
phase1 = np.unwrap(np.angle(h1))
phase2 = np.unwrap(np.angle(h2))
phase3 = np.unwrap(np.angle(h3))

# Retardo de grupo = -dφ/dω
gd1 = -np.diff(phase1) / np.diff(w1)
gd2 = -np.diff(phase2) / np.diff(w2)
gd3 = -np.diff(phase3) / np.diff(w3)

# Polos y ceros
z1, p1, k1 = signal.tf2zpk(num1, den1) # Nos devuelve la localizacion de los ceros 
z2, p2, k2 = signal.tf2zpk(num2, den2) # Nos devuelve la localizacion de los ceros 
z3, p3, k3 = signal.tf2zpk(num3, den3) # Nos devuelve la localizacion de los ceros 

# Graficos
plt.figure(figsize=(12,8))

# Magnitud
plt.subplot(2,2,1)
plt.semilogx(w1, 20*np.log10(abs(h1)), label = 'T1')
plt.semilogx(w2, 20*np.log10(abs(h2)), label = 'T2')
plt.semilogx(w3, 20*np.log10(abs(h3)), label = 'T3')
plt.title('Magnitud')
plt.xlabel('Pulsación angular [r/s]')
plt.ylabel('|H(jω)| [dB]')
plt.grid(True, which = 'both', ls = ':')
plt.legend()

# Fase
plt.subplot(2,2,2)
plt.semilogx(w1, np.degrees(phase1), label = 'T1')
plt.semilogx(w2, np.degrees(phase2), label = 'T2')
plt.semilogx(w3, np.degrees(phase3), label = 'T3')
plt.title('Fase')
plt.xlabel('Pulsación angular [r/s]')
plt.ylabel('Fase [°]')
plt.grid(True, which = 'both', ls = ':')
plt.legend()

# Retardo de grupo
plt.subplot(2,2,3)
plt.semilogx(w1[:-1], gd1, label = 'T1')
plt.semilogx(w2[:-1], gd2, label = 'T2')
plt.semilogx(w3[:-1], gd3, label = 'T3')
plt.title('Retardo de Grupo')
plt.xlabel('Pulsación angular [r/s]')
plt.ylabel('τg [s]')
plt.grid(True, which = 'both', ls = ':')
plt.legend()

# Diagrama de polos y ceros
plt.subplot(2,2,4)
plt.plot(np.real(p1), np.imag(p1), 'x', markersize = 10, label = 'Polos T1')
plt.plot(np.real(p2), np.imag(p2), 'x', markersize = 10, label = 'Polos T2')
plt.plot(np.real(p3), np.imag(p3), 'x', markersize = 10, label = 'Polos T3')

if len(z1) > 0:
    plt.plot(np.real(z1), np.imag(z1), 'o', markersize = 10, fillstyle = 'none', label = 'Ceros T1')
if len(z2) > 0:
    plt.plot(np.real(z2), np.imag(z2), 'o', markersize = 10, fillstyle = 'none', label = 'Ceros T2')
if len(z3) > 0:
    plt.plot(np.real(z3), np.imag(z3), 'o', markersize = 10, fillstyle = 'none', label = 'Ceros T3')

plt.axhline(0, color = 'k', lw = 0.5)
plt.axvline(0, color='k', lw = 0.5)
plt.title('Diagrama de Polos y Ceros (plano s)')
plt.xlabel('σ [rad/s]')
plt.ylabel('jω [rad/s]')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# III. **Conclusiones**

El analisis realizado permitio demostrar que la funcion de transferencia $T(s) = \frac{V_2(s)}{V_1(s)}$ correspondiente al circuitto propuesto puede expresarsee bajo la forma canonica de una **seccion bicuadratica (SOS)**. Esta representacion evidencia el comportamiento de un sistema de segundo orden, caracterizado por la presencia de un par de polos y un par de ceros complejos cojugados, ademas de una ganancia global $k$ que determina la magnitud de la respuesta en frecuencia. 

A partir de la parametrizacion de los polinomios de segundo orden, se logro vincular los parametros caracteristicos $\omega_{0z}$, $Q_z$, $\omega_{0p}$, $Q_p$ y $k$ con los elementos fisicos del circuito ($R$, $L$, $C$) y los coeficientes de division ($a$, $b$, $d$). Esta relacion permite interpretar directamente la influencia de cada componnte sobre la ubicacion de los polos y ceros en el plano complejo y, por lo tanto, sobre la respuesta en frecuencia. 

En cuanto a las dos secciones analizadas: 

- Presentan la misma frecuencia natural de polos $\omega_{0p} = 0.8446$, pero difieren en sus factores de calidad $Q_p$.
- El numerador proporcional a $s^2$ en ambas secciones indica la presencia de ceros dobles en el origen, caracteristica de una **respuesta de tipo pasa - alto**.

En conclusión, se verificó la correspondencia entre el modelo matemático de la función transferencia y el comportamiento físico del circuito, validando la utilidad de la representación bicuadrática como herramienta fundamental en el análisis, diseño y síntesis de sistemas lineales. 

A partir del análisis de las tres funciones de transferencia propuestas, se obtienen las siguientes observaciones:

1. $T_1(s)$:
- Los ceros se encuentran en $s = \pm j3$, mientras que los polos se ubican en $s = -\frac{\sqrt{2}}{2} \pm j\frac{\sqrt{2}}{2}$.
- La presencia de ceros en la misma banda de paso de los polos produce una anulación parcial de la señal, generando una *atenuación pronunciada* en torno a $\omega \approx 1 rad/s$.
- En el diagrama de magnitud se observa un rechazo de banda, acompañado por un cambio brusco de fase y un retardo de grupo negativo en esa región.

2. $T_2(s)$:
- Los ceros están ubicados en $s = \pm j3$, mientras que los polos se localizan en $s = -0.1 \pm j0.995$.
- Los ceros de baja frecuencia producen una fuerte atenuación en bajas frecuencias, observándose una pendiente ascendente en la magnitud.
- La fase varía desde $+180^{\circ}$ hasta $0^{\circ}$, típica de un filtro pasa-altos.
- El retardo de grupo presenta un pico acentuado en la región de transición, coherente con el cambio abrupto de fase.

3. $T_3(s)$:
- Los ceros y polos son complejos conjugados y están próximos entre sí, generando una cancelación parcial.
- La magnitud se mantiene prácticamente constante en todo el espectro, con una fase que varía suavemente.
- El retardo de grupo es prácticamente constante, indicando un comportamiento estable y de fase mínima.

En síntesis, se puede concluir que:
- $T_1(s)$ $\rightarrow$ es un **filtro de rechazo de banda**.
- $T_2(s)$ $\rightarrow$ es un **filtro pasa-altos**.
- $T_3(s)$ $\rightarrow$ es un **filtro pasa - todo**.
